# 深度模型初始化

零初始化：在第一次进行网络更新时，所有的中间层的节点的值都为0，反向传播时，更新后的网络参数将会相同；在下一次更新时，相同的网络参数学习提取不到有用的特征。收敛慢。


- 输入均匀初始化：$(- \sqrt{3}, \sqrt{3})$；
- 权重glorot初始化：$(- \sqrt{\frac{6}{fan_{in} + fan_{out}}} ,\sqrt{\frac{6}{fan_{in} + fan_{out}}} )$
- 对应不同的激活函数



网络训练冻结
- 权重非0中心；
- 数据、权重、偏差的范围太大；
- 非0中心激活函数用于深层；



# Normalization正则化

在网络合适的位置，用正确的正则化


## BatchNormalization

将数据正则化到均值为0，方差为1；再按照，scale、shift偏移。BatchNormalization 对一批样本的每个特征分别进行归一化。


## LayerNormalization

对tensor最后维度进行归一化，量纲一致。不同的特征值域等引入误差较多。在NLP中，特征较统一。























batch_norm正则化

1. 解决梯度消失：值域迁移到0附近，梯度最大
2. 解决ICS问题：各层参数分布变化大，网络加深收敛缓慢
3. 加速模型收敛：每一维数值影响相近
4. 正则化效果：正则化后引入噪声，缓解拟合
