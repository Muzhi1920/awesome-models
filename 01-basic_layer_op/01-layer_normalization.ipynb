## layer norm layer
独立地对该batch中每个样本的前一层的激活输出进行标准化，来维持每个激活输出的均值接近0，标准差接近1。

---

batch norm是跨批次的进行`Normlization`。
1. layer norm是对样本内的输出进行规范化，`axis`一般是横向；
2. batch norm是对样本外的输出，`axis`一般是跨样本的纵向；

给定张量“输入”，计算矩并归一化，在 `axis` 中指定的轴上执行。


```python
import tensorflow as tf
from tensorflow.keras.layers import Layer, Dense, LayerNormalization, Dropout, Embedding
import numpy as np
```

```python
data = tf.constant(np.arange(30).reshape(5, 6) * 10, dtype=tf.float32)
data
```

```python
layer = tf.keras.layers.LayerNormalization(axis=1)
layer
```

```python
output = layer(data)
output
```

## 结论
1. 对每一样本的输出，按照给定`axis`进行normalization
2. 而不是不同的样本间

## 引入gamma和beta
1. 在原先0，1标准正态分布基础上进行缩放和平移；
2. 初始化为0，和1，可训练。

`output_i = x_i_normalized * gamma + beta`

```python
layer = tf.keras.layers.LayerNormalization(axis=[1, 2, 3])
layer
```

```python
layer.build([5, 20, 30, 40])
layer
```
